```python
import json

# Replace 'path/to/your/file.json' with the actual path to your JSON file
with open('scatter.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

print(data)
```

```python
import numpy as np

error_x = data["data"][0]["error_x"]["array"]
error_y = data["data"][0]["error_y"]["array"]

x = data["data"][0]["x"]
y = data["data"][0]["y"]

best_fit_y = data["data"][1]["y"]
best_fit_x = data["data"][1]["x"]

mean_x = np.mean(data["data"][0]["x"])
error_x = error_x / mean_x
error_y = error_y / mean_x

x = x / mean_x
y = y / mean_x

best_fit_x = best_fit_x / mean_x
best_fit_y = best_fit_y / mean_x
```

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams

# Set global parameters to use a modern font
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Arial']

# Set a style with a white background
plt.style.use('seaborn-whitegrid')

# Create a new figure
plt.figure(figsize=(8, 6), dpi=300)

# Create a scatter plot with error bars
plt.errorbar(x, y, xerr=error_x, yerr=error_y, linestyle='None', marker='o', markersize=5, markerfacecolor='blue', alpha=0.7, markeredgecolor='blue', capsize=3, color='gray', elinewidth=1, label='Sampled Networks')
plt.plot(best_fit_x, best_fit_y, color='green', linewidth=2, label="Line of Best Fit, slope = 1.06")

# Adding labels and title with a larger font size
plt.xlabel('Normal Legs, Change in Performance Relative to Mean', fontsize=14)
plt.ylabel('Long Legs, Change in Performance Relative to Mean', fontsize=14)
plt.title('Visualising Phenotypic Diversity with Different Leg Lengths', fontsize=16)

# Increasing the size of the ticks labels and changing their colors
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Adding a grid
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.6)

# Adding a legend with a larger font size
plt.legend(fontsize=12, loc='upper left')

# Tight layout to ensure nothing is clipped
plt.tight_layout()

# Saving the figure as a high-resolution PNG file
plt.savefig('scatter_plot_with_error_bars.png', dpi=300)
```
